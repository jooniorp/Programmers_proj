```python
import pandas as pd
import numpy as np
import math
```

```python
qna_df = pd.read_excel('qna_cleaned.xlsx',sheet_name = 'qna_cleaned')
multi_qna_df = pd.read_excel('multiple_qna_cleaned.xlsx',sheet_name = 'multiple_qna_cleaned')
```

```python
qna_df.head()
```

```python
multi_qna_df.head()
```

```python
#unnamed라는 불필요한 칼럼 삭제 및 id등 트레인 데이터에 불필요한 칼럼 삭제
multi_qna_df = multi_qna_df.drop(['Unnamed: 0', 'id', 'article_id', 'menu_id'],axis = 1)
qna_df = qna_df.drop(['Unnamed: 0', 'id', 'article_id', 'menu_id'],axis = 1)
```

```python
qna_df.head()
```

```python
#[SEP]문자 제거

qna_df['question'] = qna_df['question'].str.replace("SEP", ' ').str.replace('[', '').str.replace(']','')
qna_df['answer'] = qna_df['answer'].str.replace("SEP", ' ').str.replace('[', '').str.replace(']','')
multi_qna_df['question'] = qna_df['question'].str.replace("SEP", ' ').str.replace('[', '').str.replace(']','')
multi_qna_df['answer'] = qna_df['answer'].str.replace("SEP", ' ').str.replace('[', '').str.replace(']','')
```

```python
qna_df.head()
```

```python
#mbti가 통합된 데이터 csv파일 생성
qna_df.to_csv('qna_cleaned_1.csv')
multi_qna_df.to_csv('multiple_qna_cleaned_1.csv')
```

```python
#mbti칼럼을 4개로 나눠보기

qna_df_q_mbti = qna_df['q_mbti'].to_list()
qna_df_a_mbti = qna_df['a_mbti'].to_list()
multi_qna_df_q_mbti = multi_qna_df['q_mbti'].to_list()
multi_qna_df_a_mbti = multi_qna_df['q_mbti'].to_list()

qna_df_q_mbti_IE = []
qna_df_q_mbti_NS = []
qna_df_q_mbti_TF = []
qna_df_q_mbti_PJ = []

qna_df_a_mbti_IE = []
qna_df_a_mbti_NS = []
qna_df_a_mbti_TF = []
qna_df_a_mbti_PJ = []

multi_qna_df_q_mbti_IE = []
multi_qna_df_q_mbti_NS = []
multi_qna_df_q_mbti_TF = []
multi_qna_df_q_mbti_PJ = []

multi_qna_df_a_mbti_IE = []
multi_qna_df_a_mbti_NS = []
multi_qna_df_a_mbti_TF = []
multi_qna_df_a_mbti_PJ = []

for mbti in qna_df_q_mbti:
  if type(mbti) == type(np.nan):
    qna_df_q_mbti_IE.append(np.nan)
    qna_df_q_mbti_NS.append(np.nan)
    qna_df_q_mbti_TF.append(np.nan)
    qna_df_q_mbti_PJ.append(np.nan)
  else:
    if mbti[0] == 'i' : qna_df_q_mbti_IE.append('i')
    else: qna_df_q_mbti_IE.append('e')
    if mbti[1] == 'n' : qna_df_q_mbti_NS.append('n')
    else: qna_df_q_mbti_NS.append('s')
    if mbti[2] == 'f' : qna_df_q_mbti_TF.append('f')
    else: qna_df_q_mbti_TF.append('t')
    if mbti[3] == 'p' : qna_df_q_mbti_PJ.append('p')
    else: qna_df_q_mbti_PJ.append('j')


for mbti in qna_df_a_mbti:
  if type(mbti) == type(np.nan):
    qna_df_a_mbti_IE.append(np.nan)
    qna_df_a_mbti_NS.append(np.nan)
    qna_df_a_mbti_TF.append(np.nan)
    qna_df_a_mbti_PJ.append(np.nan)
  else:
    if mbti[0] == 'i' : qna_df_a_mbti_IE.append('i')
    else: qna_df_a_mbti_IE.append('e')
    if mbti[1] == 'n' : qna_df_a_mbti_NS.append('n')
    else: qna_df_a_mbti_NS.append('s')
    if mbti[2] == 'f' : qna_df_a_mbti_TF.append('f')
    else: qna_df_a_mbti_TF.append('t')
    if mbti[3] == 'p' : qna_df_a_mbti_PJ.append('p')
    else: qna_df_a_mbti_PJ.append('j')


for mbti in multi_qna_df_q_mbti:
  if type(mbti) == type(np.nan):
    multi_qna_df_q_mbti_IE.append(np.nan)
    multi_qna_df_q_mbti_NS.append(np.nan)
    multi_qna_df_q_mbti_TF.append(np.nan)
    multi_qna_df_q_mbti_PJ.append(np.nan)
  else:
    if mbti[0] == 'i' : multi_qna_df_q_mbti_IE.append('i')
    else: multi_qna_df_q_mbti_IE.append('e')
    if mbti[1] == 'n' : multi_qna_df_q_mbti_NS.append('n')
    else: multi_qna_df_q_mbti_NS.append('s')
    if mbti[2] == 'f' : multi_qna_df_q_mbti_TF.append('f')
    else: multi_qna_df_q_mbti_TF.append('t')
    if mbti[3] == 'p' : multi_qna_df_q_mbti_PJ.append('p')
    else: multi_qna_df_q_mbti_PJ.append('j')

for mbti in multi_qna_df_a_mbti:
  if type(mbti) == type(np.nan):
    multi_qna_df_a_mbti_IE.append(np.nan)
    multi_qna_df_a_mbti_NS.append(np.nan)
    multi_qna_df_a_mbti_TF.append(np.nan)
    multi_qna_df_a_mbti_PJ.append(np.nan)
  else:
    if mbti[0] == 'i' : multi_qna_df_a_mbti_IE.append('i')
    else: multi_qna_df_a_mbti_IE.append('e')
    if mbti[1] == 'n' : multi_qna_df_a_mbti_NS.append('n')
    else: multi_qna_df_a_mbti_NS.append('s')
    if mbti[2] == 'f' : multi_qna_df_a_mbti_TF.append('f')
    else: multi_qna_df_a_mbti_TF.append('t')
    if mbti[3] == 'p' : multi_qna_df_a_mbti_PJ.append('p')
    else: multi_qna_df_a_mbti_PJ.append('j')


print(len(qna_df_q_mbti_IE))
print(len(qna_df_q_mbti_NS))
print(len(qna_df_q_mbti))
```
