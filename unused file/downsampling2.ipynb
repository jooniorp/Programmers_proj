```python
df1=df['mbti']=="intp"
df2=df['mbti']=="intj"
df3=df['mbti']=="infp"
df4=df['mbti']=="infj"
df5=df['mbti']=="istp"
df6=df['mbti']=="istj"
df7=df['mbti']=="isfp"
df8=df['mbti']=="isfj"
df9=df['mbti']=="entp"
df10=df['mbti']=="entj"
df11=df['mbti']=="enfp"
df12=df['mbti']=="enfj"
df13=df['mbti']=="estp"
df14=df['mbti']=="estj"
df15=df['mbti']=="esfp"
df16=df['mbti']=="esfj"

```

```python
df_intp=df[df1]
df_intj=df[df2]
df_infp=df[df3]
df_infj=df[df4]
df_istp=df[df5]
df_istj=df[df6]
df_isfp=df[df7]
df_isfj=df[df8]
df_entp=df[df9]
df_entj=df[df10]
df_enfp=df[df11]
df_enfj=df[df12]
df_estp=df[df13]
df_estj=df[df14]
df_esfp=df[df15]
df_esfj=df[df16]
```

```python
df_intp.head(5)
```

```python
df_nf=pd.concat([df_infp,df_infj,df_enfp,df_enfj])
```

```python
df_nf['text'].count()
```

```python
df_nt=pd.concat([df_intp,df_intj,df_entp,df_entj])
df_sf=pd.concat([df_isfp,df_isfj,df_esfp,df_esfj])
df_st=pd.concat([df_istp,df_istj,df_estp,df_estj])
```

```python
df_nt['text'].count(),df_sf['text'].count(),df_st['text'].count()
```

```python
df_nf = df_nf.sample(frac=1).reset_index(drop=True)
df_nt = df_nt.sample(frac=1).reset_index(drop=True)
df_sf = df_sf.sample(frac=1).reset_index(drop=True)
df_st = df_st.sample(frac=1).reset_index(drop=True)
```

```python
df_nf=df_nf.sample(n=18322)
df_nt=df_nt.sample(n=18322)
df_sf=df_sf.sample(n=18322)
```

```python
df_nf['text'].count(),df_nt['text'].count(),df_sf['text'].count()
```

```python
df_nf.head(5)
```

```python
df1=pd.concat([df_nt,df_nf,df_st,df_sf])
df1 = df1.sample(frac=1).reset_index(drop=True)
df1.head(10)
```

```python
df1['text'].count()
```

```python
df1 = df1.replace({'mbti' : 'intp'}, 'nt')
df1 = df1.replace({'mbti' : 'intj'}, 'nt')
df1 = df1.replace({'mbti' : 'infp'}, 'nf')
df1 = df1.replace({'mbti' : 'infj'}, 'nf')
df1 = df1.replace({'mbti' : 'istp'}, 'st')
df1 = df1.replace({'mbti' : 'istj'}, 'st')
df1 = df1.replace({'mbti' : 'isfp'}, 'sf')
df1 = df1.replace({'mbti' : 'isfj'}, 'sf')

df1 = df1.replace({'mbti' : 'entp'}, 'nt')
df1 = df1.replace({'mbti' : 'entj'}, 'nt')
df1 = df1.replace({'mbti' : 'enfp'}, 'nf')
df1 = df1.replace({'mbti' : 'enfj'}, 'nf')
df1 = df1.replace({'mbti' : 'estp'}, 'st')
df1 = df1.replace({'mbti' : 'estj'}, 'st')
df1 = df1.replace({'mbti' : 'esfp'}, 'sf')
df1 = df1.replace({'mbti' : 'esfj'}, 'sf')
```

```python
df1.to_csv('data_set\cafe_data\df_nf_4.csv')
```
