```python
#데이터 형식 통일 text열, mbti열 mbti는 소문자로
df_ka['mbti'] = df_ka['MBTI'].str.lower()
df_ka.drop(['MBTI','Unnamed: 0'], axis = 1, inplace = True)
df_ka.head()
```

```python
df2['mbti'] = df2['mbti'].str.lower()
df3['mbti'] = df3['mbti'].str.lower()
df4['mbti'] = df4['mbti'].str.lower()
df5['mbti'] = df5['mbti'].str.lower()
df6['mbti'] = df6['mbti'].str.lower()
df7['mbti'] = df7['mbti'].str.lower()
```

```python
df2 = df2[['text','mbti']]
df3 = df3[['text','mbti']]
df4 = df4[['text','mbti']]
df5 = df5[['text','mbti']]
df6 = df6[['text','mbti']]
df7 = df7[['text','mbti']]
```
